```python
import pandas as pd
data=pd.read_csv("Advertising.csv")
```

```python
data
```

```python
data.info()
```

```python
x=data.drop(['Unnamed: 0','sales'],axis=1)
y=data['sales']
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=123)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
prediction=model.predict(x_test)
prediction
```

```python
df=pd.DataFrame(prediction)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error,mean_absolute_percentage_error
```

```python
mean_squared_error(y_test,prediction)
```

```python
r2_score(y_test,prediction)
```

```python
mean_absolute_error(y_test,prediction)
```

```python
mean_absolute_percentage_error(y_test,prediction)
```

```python
import numpy as np
```

```python
from sklearn.preprocessing import PolynomialFeatures
for i in range(1,10):
    poly_features=PolynomialFeatures(degree=i)
    x_poly=poly_features.fit_transform(x_train)
    x_test_poly=poly_features.fit_transform(x_test)
    model=LinearRegression()
    model.fit(x_poly,y_train)
    final=model.predict(x_test_poly)
    print('For Degree ', i)
    print('mean_squared_error=',mean_squared_error(y_test,final))
    print('r2_score=',r2_score(y_test,final))
    print('mean_absolute_error=',mean_absolute_error(y_test,final))
    print('mean_absolute_percentage_error=',mean_absolute_percentage_error(y_test,final))
    
```

```python
poly_features=PolynomialFeatures(degree=2)
x_poly=poly_features.fit_transform(x_train)
x_test_poly=poly_features.fit_transform(x_test)
model=LinearRegression()
model.fit(x_poly,y_train)
final=model.predict(x_test_poly)
```
