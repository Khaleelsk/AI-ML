```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
z = np.array([10,20,30,40])
plt.plot(x,y,z)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Plot 1')
plt.xlabel('X1')
plt.ylabel('Y1')

plt.subplot(1,2,2)
plt.plot(x,z)
plt.title('Plot 2')
plt.xlabel('X2')
plt.ylabel('Y2')

plt.suptitle('Plots')

plt.tight_layout()

plt.show()
```
