```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url='https://www.makaan.com/vijayawada-residential-property/buy-property-in-vijayawada-city?page='
```

```python
Flat_Type=[]
Location=[]
Final_Prices=[]
SFT=[]
SellerInfo=[]
Status=[]
Construction=[]
```

```python
for i in range(1,15):
    url += str(i)
    data = requests.get(url)
    soup = BeautifulSoup(data.content, 'html')
    
    #flattype
    a=soup.find_all('a',attrs={'typelink'})
    flattype=[i.text for i in a]
    Flat_Type+=flattype
    
    #location
    b=soup.find_all('a',attrs={'class':'loclink'})
    loc=[i.text for i in b]
    Location+=loc
    
    #prices
    c=soup.find_all('td', attrs={'class' : 'price'})
    prices=[i.text for i in c]
    Final_Prices+= prices
    
    #sft
    d=soup.find_all('td',attrs={'class':'size'})
    sft=[i.text for i in d]
    SFT+=sft
    
    #status
    e=soup.find_all('td',attrs={'class':'val'})
    status=[i.text for i in e]
    Status+=status
    
    #construction & Others
    f=soup.find_all('ul',attrs={'class':'listing-details'})
    constr=[i.text for i in f]
    Construction+=constr
    
    #sellerinfo
    g=soup.find_all('div',attrs={'class':'seller-info'})
    sellerinfo=[i.span.text for i in g]
    SellerInfo+=sellerinfo
    
    url=url[:-1]
    
    #convertion of Prices
    final_prices= []
    for price in Final_Prices:
        if price.find('Cr')!=-1:
            final_prices.append(float(price.replace('Cr', '')) * 10000000)
        else:
            final_prices.append(float(price.replace('L', '')) * 100000)
    myfile={
        "Title":Flat_Type,
        "Price":final_prices,
        "Location":Location,
        "Area(SFT)":SFT,
        "Seller_Info":SellerInfo,
        "Construction & Others":Construction,
        "Status":Status
    }
myfile
```

```python
DF=pd.DataFrame(myfile)
```

```python
DF.to_csv("Makaan_Web_Scraping.csv")
```
