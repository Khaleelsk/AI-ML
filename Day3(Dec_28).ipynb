```python
india_data=px.data.gapminder().query("country=='India'")
india_data
```

```python
gapminder_2007=gapminder.query("year==2007")
gapminder_2007
```

```python
fig=px.scatter(gapminder_2007,x='gdpPercap',y='lifeExp',color='continent',height=350)
fig.show()
```

```python
fig=px.scatter(gapminder_2007,x='gdpPercap',y='lifeExp',color='continent',hover_name='country',size='pop')
fig.show()
```

```python
fig=px.scatter(gapminder_2007,
              x='gdpPercap',
              y='lifeExp',
              color='continent',
              size='pop',
              size_max=60,
              facet_col='continent',log_x=True)
fig.show()
```

```python
fig=px.scatter(gapminder_2007,
              x='gdpPercap',hover_name='country',
              y='lifeExp',
              color='lifeExp',
              size='pop',
              size_max=60,
              facet_col='continent',
               range_x=[25,50000],
               range_y=[25,90],
               log_x=True)
fig.show()
```

```python
fig=px.scatter(gapminder,
              x='gdpPercap',hover_name='country',
              y='lifeExp',
              color='continent',
              size='pop',
              size_max=60,
              facet_col='continent',
               range_x=[25,10000],
               range_y=[25,90],
               animation_frame='year',
               animation_group='country',
               log_x=True)
fig.show()
```

```python
fig=px.bar(gapminder,
              x='gdpPercap',hover_name='country',
              y='lifeExp',
              color='lifeExp',
              facet_col='continent',
               animation_frame='year',
               animation_group='country',
               log_x=True)
fig.show()
```

```python
fig=px.choropleth(gapminder,locations='iso_alpha',color='lifeExp',hover_name='country',animation_frame='year',color_continuous_scale=px.colors.sequential.Plasma,
                 projection='natural earth')
fig.show()
```
